```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_excel("./data/망고플레이트 리뷰 종합.xlsx")
```

```python
# pandas dataframe to SQLite3

con = sqlite3.connect("mango_review.db")

df.to_sql("mango_plate", con, if_exists="append", index=False)

con.close()
```

```python
# pandas SQLite3 to dataframe

con = sqlite3.connect("mango_review.db")

df = pd.read_sql("SELECT * FROM mango_plate", con)

con.close()
```
