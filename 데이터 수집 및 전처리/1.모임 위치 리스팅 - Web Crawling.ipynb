```python
gu=["광진구", "구로구", "금천구", "노원구", "도봉구"]
baseurl="https://www.mangoplate.com/"

name=[]
score=[]
address=[]
service=[]
price=[]
urllist=[]

for i in range(len(gu)):
    guname=gu[i]
    print("-"*80)
    for j in range(10):
        url = baseurl + "search/" + guname + "?keyword=" +guname + "&page=" + str(j+1)
        
        headers={"User-Agent":"Mozilla/5.0"}
        page=requests.get(url, headers=headers)
        page

        soup=BeautifulSoup(page.text, "html.parser") #request라는 모듈로 가져왔을 시에는 text 가져와야 함
        soup

        base=soup.find_all("div", "info")
        for k in range(20):
            one=base[k]
            url3="https://www.mangoplate.com/"+one.a["href"]
            urllist.append(url3)
            
#             url2=one.a["href"]
#             base= "https://www.mangoplate.com/"

#             url3=base+url2

            headers={"User-Agent":"Mozilla/5.0"}
            page=requests.get(url3, headers=headers)
            page

            soup=BeautifulSoup(page.text, "html.parser") #request라는 모듈로 가져왔을 시에는 text 가져와야 함
            soup

            n=soup.find_all("h1")[0]
            name.append(n.get_text())
            
            sc=soup.find_all("strong")[0].get_text()
            score.append(sc)
            
            a=soup.find_all("td")[0]
            address.append(a.get_text().split("\n")[0])
            
            s=soup.find_all("td")[2]
            service.append(s.get_text())
            
            try:
                p=soup.find_all("td")[3].get_text()
            except:
                p=""
            price.append(p)
```
