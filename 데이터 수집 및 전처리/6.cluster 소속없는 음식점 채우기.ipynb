```python
for i in idx:
    print(i)
    distance = []
    location = (places["위도"][i], places["경도"][i])
    for j in range(len(clusters)):
        d=haversine(location, (clusters["위도"][j], clusters["경도"][j]))
        distance.append(d)
    min_distance_point=distance.index(min(distance))
    places["Cluster"][i] = [min_distance_point]
```
